
# Merge the 3 signers videos 

```python
import os
import shutil

# Define the main folders
main_folder_01 = "D:/Year 3/GP/GP100Word/Arabic_ActionDetectionforSignLanguage(100)/dataset/01"
main_folder_02 = "D:/Year 3/GP/GP100Word/Arabic_ActionDetectionforSignLanguage(100)/dataset/02"
main_folder_03 = "D:/Year 3/GP/GP100Word/Arabic_ActionDetectionforSignLanguage(100)/dataset/03"

# Get the list of subfolders in main_folder_01/train
subfolders_01 = os.listdir(os.path.join(main_folder_01, "train"))

# Iterate over the subfolders in main_folder_01/train
for subfolder in subfolders_01:
    # Create the source and destination folder paths
    source_folder_02 = os.path.join(main_folder_02, "train", subfolder)
    source_folder_03 = os.path.join(main_folder_03, "train", subfolder)
    destination_folder_01 = os.path.join(main_folder_01, "train", subfolder)
    
    # Check if the source folders exist
    if os.path.exists(source_folder_02) and os.path.exists(source_folder_03):
        # Get the list of files in source_folder_02
        files_02 = os.listdir(source_folder_02)
        
        # Iterate over the files in source_folder_02
        for file_name in files_02:
            # Create the source and destination file paths
            source_file_02 = os.path.join(source_folder_02, file_name)
            destination_file_01 = os.path.join(destination_folder_01, file_name)
            
            # Copy the file from source_folder_02 to destination_folder_01
            shutil.copy(source_file_02, destination_file_01)
            print(f"File '{file_name}' copied to '{destination_folder_01}'")
            
        # Get the list of files in source_folder_03
        files_03 = os.listdir(source_folder_03)
        
        # Iterate over the files in source_folder_03
        for file_name in files_03:
            # Create the source and destination file paths
            source_file_03 = os.path.join(source_folder_03, file_name)
            destination_file_01 = os.path.join(destination_folder_01, file_name)
            
            # Check if the file already exists in destination_folder_01
            if not os.path.exists(destination_file_01):
                # Copy the file from source_folder_03 to destination_folder_01
                shutil.copy(source_file_03, destination_file_01)
                print(f"File '{file_name}' copied to '{destination_folder_01}'")
```

```python
import pandas as pd
import os

# Folder path containing subfolders with SignID names
folder_path = 'D:/Year 3/GP/GP100Word/Arabic_ActionDetectionforSignLanguage(100)/dataset/01/train'

# Read the Excel file
excel_file = 'D:/Year 3/GP/GP100Word/Arabic_ActionDetectionforSignLanguage(100)/KARSL-100_Labels.xlsx'
df = pd.read_excel(excel_file)

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Retrieve the SignID and Sign-English values
    sign_id = str(row['SignID'])
    sign_english = str(row['Sign-English'])
    
    # Construct the full path to the folder

    folder_to_rename = os.path.join(folder_path,sign_id)
    
    # Check if the folder exists
    if os.path.isdir(folder_to_rename):
        # Rename the folder
        new_folder_name = os.path.join(folder_path, sign_english)
        os.rename(folder_to_rename, new_folder_name)
        print(f"Renamed folder '{folder_to_rename}' to '{new_folder_name}'")
    else:
        print(f"Folder '{folder_to_rename}' does not exist.")
```
